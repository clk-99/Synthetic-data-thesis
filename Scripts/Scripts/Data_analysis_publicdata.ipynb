## Bank marketing data

```python
bank_df = pd.read_csv(data_folder + '/BANK_MARKETING/BANK_TRAIN_SET.csv',delimiter=',',index_col=0)
bank_df
```

```python
bank_temp = bank_df.copy()
```

```python
table_dict = dict()
table_dict['METADATA_SPEC_VERSION'] = 'SINGLE_TABLE_V1'
object_columns = bank_df.dtypes[bank_df.dtypes=='object'].index.to_list()
int_columns = bank_df.dtypes[bank_df.dtypes=='int64'].index.to_list()
float_columns = bank_df.dtypes[bank_df.dtypes=='float64'].index.to_list()
all_columns = bank_df.dtypes.index.to_list()
col_dict = dict()

for col in all_columns:
    #create for each column a dictionary and add to table_dict
    if col in float_columns:
        float_dict = dict()
        float_dict['sdtype'] = 'numerical'
        float_dict['computer_representation'] = 'Float'
        col_dict[col] = float_dict

    elif col in int_columns:
        int_dict = dict()
        int_dict['sdtype'] = 'numerical'
        int_dict['computer_representation'] = 'Int64'
        col_dict[col] = int_dict
    
    else: #col in object_columns
        object_dict = dict()
        object_dict['sdtype'] = 'categorical'
        col_dict[col] = object_dict

table_dict['columns'] = col_dict
table_dict    
```
